```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('plotting_function.csv')  # Replace with your file

# a. Normal Curve
sns.kdeplot(df['Glucose'])
plt.title("Normal Curve for Glucose")
plt.show()

# b. Density and Contour Plots
sns.kdeplot(x=df['Glucose'], y=df['BMI'])
plt.title("Density Contour Plot")
plt.show()

# c. Correlation and Scatter
sns.scatterplot(x=df['Age'], y=df['Glucose'])
plt.title("Scatter Plot")
plt.show()

# d. Histogram
df['BMI'].hist()
plt.title("Histogram of BMI")
plt.show()

# e. 3D Plotting
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Glucose'], df['BMI'], df['Age'])
ax.set_xlabel('Glucose')
ax.set_ylabel('BMI')
ax.set_zlabel('Age')
plt.show()
```
